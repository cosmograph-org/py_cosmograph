# Visualizing the Lexical Cosmos: Exploring Word Relationships with Cosmograph

```python
# Install Cosmograph
# %pip install --quiet cosmograph
```

```python
import pandas as pd
from cosmograph import cosmo
```

## 1. Mapping the Lexical Landscape

```python
df = pd.read_parquet('https://www.dropbox.com/scl/fi/4mnk1e2wx31j9mdsjzecy/wordnet_feature_meta.parquet?rlkey=ixjiiso80s1uk4yhx1v38ekhm&dl=1')
print(f"{df.shape=}")
df.iloc[0]
```

```python
cosmo(
    df,
    point_id_by='lemma',
    point_label_by='word',
    point_x_by='umap_x',
    point_y_by='umap_y',
    point_color_by='frequency',
    point_size_by='frequency',
    point_size_scale=3,
)
```

## 2. Revealing Hierarchical Relationships

```python
hyponyms = pd.read_parquet('https://www.dropbox.com/scl/fi/pl72ixv34soo1o8zanfrz/hyponyms.parquet?rlkey=t4d606fmq1uinn29qmli7bx6r&dl=1')
print(f"{hyponyms.shape=}")
hyponyms.iloc[0]
```

```python
cosmo(
    points=df,
    links=hyponyms,
    link_source_by='source',
    link_target_by='target',
    point_id_by='lemma',
    point_label_by='word',
    # point_x_by='umap_x',
    # point_y_by='umap_y',
    point_color_by='frequency',
    point_size_by='frequency',
    point_size_scale=2,
)
```
