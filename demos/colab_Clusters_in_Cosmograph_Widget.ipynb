# Visualizing 10,000 Data Points with Cosmograph Widget: A Cluster Analysis

```python
# Install Cosmograph Widget
# !pip install --quiet cosmograph
```

```python
from cosmograph import cosmo
import numpy as np
import pandas as pd
```

```python
# Generate 10,000 data points with 10 distinct clusters

clusters = np.array([i % 10 for i in range(10000)])
points = pd.DataFrame({ 'id': range(10000), 'cluster': np.array([f"Cluster {x}" for x in clusters]) })

print(f"{points.shape=}")
points.iloc[0]
```

```python
cosmo(
   points,
   point_id_by='id',
   point_cluster_by='cluster', # Automatic cluster detection
   point_color_by='cluster',   # Color points by their cluster

   show_cluster_labels=True,
   use_point_color_strategy_for_cluster_labels=True,
   select_cluster_on_label_click=True,
)
```
