```python
from collections import defaultdict

def find_duplicates(pairs, *, val_filt=lambda x: len(x) > 1):
    d = defaultdict(list)
    for key, value in pairs:
        d[key].append(value)
    return {k: v for k, v in d.items() if val_filt(v)}
```

```python
import json 
from operator import itemgetter

from cosmograph.util import data_files

def to_dict(iterable, key_field: str):
    extract_key = itemgetter(key_field)
    return {extract_key(x): x for x in iterable}

# config_info
config_info = json.loads(data_files['config_info.json'])
config_info_properties = to_dict(config_info, 'Property') #{x['Property']: x for x in config_info}
assert len(config_info_properties) == len(config_info), f"config_info had some duplicate properties"


# cosmos_config
cosmos_config = json.loads(data_files['cosmos_config.json'])
assert list(cosmos_config) == ['config']
cosmos_config = cosmos_config['config']
cosmo_config_properties = to_dict(cosmos_config, 'Property') 
assert len(cosmo_config_properties) == len(cosmos_config), f"cosmos_config had some duplicate properties"


# widget_config_from_md
widget_config_from_md = json.loads(data_files['_widget_config_from_md.json'])
assert set(widget_config_from_md) == {'interfaces'}
assert len(widget_config_from_md['interfaces']) == 1
assert set(widget_config_from_md['interfaces'][0]) == {'description', 'name', 'properties'}

def remove_empty_defaults(elements):
    for d in elements:
        if d['default'] == 'NO_DEFAULT':
            d.pop('default')
        yield d

t = remove_empty_defaults(widget_config_from_md['interfaces'][0]['properties'])
widget_config_from_md_properties = to_dict(t, 'name')

assert len(widget_config_from_md_properties) == len(widget_config_from_md['interfaces'][0]['properties']), f"widget_config_from_md had some duplicate properties"


```

```python
# widget_config
widget_config = json.loads(data_files['_widget_config.json'])


from lkj import camel_to_snake


def _widget_config_properties():
    assert sorted(widget_config) == [
        'config.ts',
        'cosmos/config.ts',
        'data.ts',
        'labels.ts',
        'simulation.ts',
    ]

    for widget_config_value in widget_config.values():
        assert set(widget_config_value) == {'interfaces'}
        interfaces = widget_config_value['interfaces']
        for interface in interfaces:
            assert set(interface) == {'description', 'name', 'properties'}
            for prop in interface['properties']:
                assert set(prop).issuperset(
                    {'description', 'name', 'type'}
                ), f"missing fields: {set(['description', 'name', 'type']) - set(prop)=})"
                prop['py_name'] = camel_to_snake(prop['name'])
                prop['origin_name'] = interface['name']
                yield prop


import i2

interfaces = list(_widget_config_properties())
print(f"{len(interfaces)=}")

widget_config_properties = to_dict(interfaces, 'name')
# assert len(widget_config_properties) == len(interfaces), f"widget_config had some duplicate properties"

duplicates = find_duplicates(((x['name'], x) for x in interfaces))
assert list(duplicates) == [
    'disableSimulation',
    'showTopLabelsLimit',
    'showHoveredPointLabel',
    'staticLabelWeight',
    'dynamicLabelWeight',
    'labelMargin',
    'labelPadding',
]

# Solve all duplicates taking the CosmographConfig orgin_name



def dedupper(dups):
    if len(dups) == 1:
        return dups[0]
    else:
        return next(filter(lambda x: x['origin_name'] == 'CosmographConfig', dups))
    
dedupped_interfaces = find_duplicates(
    ((x['name'], x) for x in interfaces), val_filt=lambda x: True
)
dedupped_interfaces = list(map(dedupper, dedupped_interfaces.values()))

widget_config_properties = to_dict(dedupped_interfaces, 'name')

assert len(widget_config_properties) == len(dedupped_interfaces), f"widget_config had some duplicate properties"
print(f"{len(widget_config_properties)=}")

```

```python
def properties_to_doc_string(props):
    """Make a doc string from a list of properties"""
    return '\t' + '\n\t'.join(f"{prop['name']}: {prop['description']}" for prop in props.values())

properties_to_doc_string(configs)
```

```python
def preprocess_points_and_links(points, links):
    # assert links is None, f"links was {links}"
    if isinstance(points, pd.DataFrame):
        points = points.copy()
        if 'index' not in points.columns:
            points['index'] = range(len(points))

    return points, links 

def add_to_doc(func):
    if not func.__doc__:
        func.__doc__ = ''
    func.__doc__ += f'Args:\n{properties_to_doc_string(configs)}\n'
    return func

def process_inputs(points, links, **cosmo_configs):
    points, links = preprocess_points_and_links(points, links)
    if 'point_x' not in cosmo_configs and 'x' in points.columns:
        cosmo_configs['point_x'] = 'x'
    if 'point_y' not in cosmo_configs and 'y' in points.columns:
        cosmo_configs['point_y'] = 'y'
    return points, links, cosmo_configs
    
@add_to_doc
@new_sig.inject_into_keyword_variadic
def cosmo(points, links=None, **cosmo_configs):
    try:
        points, links, cosmo_configs = process_inputs(points, links)
        return Cosmograph(points=points, links=links, **cosmo_configs)
    except Exception as e:
        print(f"Error: {e}")
        print(f"{points=}, {links=}, {cosmo_configs=}")
        raise e

help(cosmo)
```
